```python
from docx import Document
import pandas as pd
from IPython.display import display
```

```python
# Adding the FAQ doc which has the format in QA and T 

# The tags used here will be for example: 
# 1. Grant 2. Scholarships 3. Proposal 4. Research 5. Policies 6. Other 7.Unknown

doc_path = 'New Faculty Orientation Research FAQs 2021 (1).docx' 
doc = Document(doc_path)
faq_df = doc2df(doc_path)
```

```python
# Function to convert doc to a dataframe

def doc2df(doc_path):

    # Using dictionary data structure
    data = {'Question': [], 'Answer': [], 'Tag': []}


    # Parsing over paragraphs and extracting Question, Answer and Tags info
    for i, para in enumerate(doc.paragraphs):
        text = para.text.strip()
        if text.startswith('Question:'):
            data['Question'].append(text[len('Question:'):].strip())
        elif text.startswith('Answer:'):
            data['Answer'].append(text[len('Answer:'):].strip())
        elif text.startswith('Tag:'):
            data['Tag'].append(text[len('Tag:'):].strip())

    # Creating DataFrame from the collected data
    faq_df = pd.DataFrame(data)

    return faq_df
```

```python
# Printing no of questions present in the document
print(f'Number of questions: {len(faq_df)}')

# Displaying the DataFrame as a table
display(faq_df)
```
